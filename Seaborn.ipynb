**SEA BORN**

```python
import seaborn as sns
import pandas as pd
```

**DATA**

```python
tips= pd.read_excel('/content/Tips (1).xlsx')
```

```python
tips
```

```python
tips.head()
```

```python
tips=tips.set_index('Unnamed: 0')  #to set the index select the name of column
```

```python
tips
```

```python
#Displot (distribution plot) histogram

sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'],kde=True,bins=30)  #bins means frequency of distribution and KDE mean the following line
```

**Joint plot**

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') #define and x and y axis source of file and kind #joint plot give compare two distribution plots
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')  #a better understanding how much dark your value, thats a hotspot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

**Pairplot **

```python
sns.pairplot(tips)  #compare all numericals columns
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm') #giving same comparision but catogorical like male and female
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

**Categorical data plots**

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips) #x,y and column
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

VIOLIN PLOT

```python
sns.violinplot(x='day', y='total_bill',data=tips,palette='rainbow')
#distribution of total bill per day
```

```python
sns.violinplot(x='day', y='total_bill',data=tips, hue='sex',palette='Set1')
```

```python
sns.violinplot(x='day', y='total_bill',data=tips, hue='sex',palette='Set3',split='True')
```

Strip plot

```python
sns.stripplot(x="day",y='total_bill',data=tips,palette='Set2')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```
