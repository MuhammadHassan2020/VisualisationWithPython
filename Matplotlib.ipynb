```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Plotting with Numpy Arrays
x= np.linspace(0,5,11)
y= x**2
x
```

```python
plt.plot(x,y)
```

```python
#tunning of graph
plt.plot(x,y,color='green',marker='o',linestyle='dashed',linewidth=1,markersize=1)
```

```python
#TITLE AND LABELS ARE VERY IMPORTANT
plt.plot(x,y,color='green',marker='o',linestyle='dashed',linewidth=1,markersize=1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('LSBU Guys practicing')
plt.show()
```

```python
#we are using this like when we are using loop to updating i on every iteration
i=2
plt.plot(x, 4+x**i,'g')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('even more practice')
plt.show()
```

```python
#Creating Multiplots on same canvas => plt.subplot(nrows, ncols, plot_number)
plt.subplot(2,4,1)  #2 is for rows and 4 is for column and 1 is for the position
plt.plot(x,y,'r--')
plt.subplot(2,2,4)
plt.plot(x,x**2,'g*-')
plt.subplot(2,2,2)
plt.plot(x,x**3,'g*-')
plt.subplot(2,2,3)
plt.plot(x,x**6,'g*-')
```

```python
# use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

#Now use the axes object to add stuff to plot
axes.plot(x,x**4,'o')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('bla bla')
```

```python
# you can specify the spubplots to draw subplots on nrows and ncols
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python

for ax in axes:
  ax.plot(x,y,'b')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('title')
#display the figure object
fig
```

```python
fig, axes = plt.subplots(figsize=(30,10),nrows=1,ncols=6)
i=0
for ax in axes:
  ax.plot(x,x**i,'g')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('power' + str(i))
  i+=1

fig
plt.tight_layout()
```
